```python
# Import the libraries.
import cv2  
import numpy as np  
import random
import matplotlib.pyplot as plt
```

### S1c. Number of teeth from a gear wheel.

```python
# Read the input image.
img = cv2.imread(r'./Input Images/wheel.png', 0)  

# Define the kernel.
kernel = np.ones((3,3), np.uint8)  

def GrassAlgorithm(image, conn):
    zone = 0
    # Threshold operation for the black pixels.
    _, imageTh = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
    # Gray Image to Color Image.
    out = cv2.cvtColor(imageTh, cv2.COLOR_GRAY2BGR)
    height, width = imageTh.shape
    # In order to manage the borders, the mask is +2 pixels bigger in each side. 
    mask = np.zeros((height+2, width+2), np.uint8)
    for i in range(height):
        for j in range(width):
            if(imageTh[i,j] == 255):
                zone = zone + 1
                # Generate random color for the region.
                r = random.randint(0,255)
                g = random.randint(0,255)
                b = random.randint(0,255)
                # Identify the connections and assign the random color.
                cv2.floodFill(out, mask, (j,i), (b,g,r), flags = conn)
    return out, zone

def Zones(img,out):
    _, image = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    height, width = image.shape
    colors = []
    for i in range(height):
        for j in range(width):
            if(out[i,j] not in colors):
                colors.append(img[i,j])
    return len(colors)

# Operations.
img_closing = cv2.dilate(cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=6), kernel, iterations=1)
img_opening = cv2.erode(cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=1), kernel, iterations=1)
img_subs = img_closing - img_opening
img_erosion = cv2.morphologyEx(img_subs, cv2.MORPH_OPEN, kernel, iterations=3)
#cv2.erode(cv2.morphologyEx(img_subs, cv2.MORPH_OPEN, kernel, iterations=2), kernel, iterations=2)
#cv2.morphologyEx(cv2.erode(img_subs, kernel, iterations=3), cv2.MORPH_OPEN, kernel, iterations=1)
grass = GrassAlgorithm(img_erosion, 4)

plt.figure(figsize=(17,17))
plt.subplot(3,3,1)
plt.title("Original image.")
plt.imshow(img,cmap = 'gray');

plt.subplot(3,3,2)
plt.title("Opening1 image.")
plt.imshow(img_opening,cmap = 'gray')

plt.subplot(3,3,3)
plt.title("Closing5 image.")
plt.imshow(img_closing,cmap = 'gray')

plt.subplot(3,3,4)
plt.title("Closing5 - Opening1")
plt.imshow(img_subs,cmap = 'gray')

plt.subplot(3,3,5)
plt.title("Erosion(Closing5 - Opening1)")
plt.imshow(img_erosion,cmap = 'gray')

plt.subplot(3,3,6)
plt.title("Grassfire(Erosion(Closing5 - Opening1))")
plt.imshow(grass[0],cmap = 'gray')

print("The wheel has "+ str(grass[1]) + " pixels in the teeth.")
print("The wheel has "+ str(Zones(img,grass[0])) + " teeth.")
#plt.savefig('./Output Images/S1c Wheel.png')
```
